# Informe Final: Análisis de Evasión de Clientes (Churn)

La evasión de clientes o churn es un problema relevante para cualquier empresa de telecomunicaciones porque impacta directamente en los ingresos y en la estabilidad del negocio. El objetivo de este análisis fue comprender cómo se comportan los clientes, qué factores influyen en la cancelación del servicio y qué medidas se pueden tomar para reducirla.

Para comenzar, se importaron los datos desde un archivo JSON. Fue necesario normalizar la estructura porque tenía información anidada en varias columnas. Durante el proceso se detectaron valores faltantes en la variable churn que fueron limpiados. También se creó una nueva variable llamada Cuentas_Diarias, calculada a partir del cargo mensual dividido en treinta días, con el fin de tener una visión más detallada del gasto diario de los clientes.

El análisis exploratorio mostró que la mayoría de los clientes permanecen en la compañía, pero existe un grupo importante que cancela. Se observó que los clientes con contrato mes a mes presentan una tasa de evasión mucho más alta que aquellos con contratos de uno o dos años. También se detectó que los clientes que utilizan el método de pago Electronic check tienen mayor tendencia a abandonar el servicio. En cuanto a las variables numéricas, los clientes con pocos meses de permanencia son los que más cancelan y aquellos con cargos mensuales altos también tienen más riesgo de irse.

A partir de estos hallazgos se puede concluir que la evasión está asociada con la antigüedad del cliente, el tipo de contrato y el método de pago. Los clientes nuevos, los que pagan con medios electrónicos inmediatos y los que enfrentan facturas altas son más vulnerables a abandonar la compañía.

Finalmente, se recomienda incentivar contratos de mayor duración ofreciendo beneficios o descuentos, mejorar la experiencia de quienes usan Electronic check mediante la promoción de métodos de pago alternativos, diseñar ofertas personalizadas para clientes con cargos mensuales altos y aplicar programas de fidelización en los primeros meses de servicio. Estas acciones ayudarían a reducir la evasión y a fortalecer la relación con los clientes.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

df_proc = df.copy()
df_proc["Churn"] = df_proc["Churn"].replace("", pd.NA)
df_proc = df_proc.dropna(subset=["Churn"])
df_proc["Churn_num"] = df_proc["Churn"].map({"No": 0, "Yes": 1})

if "customerID" in df_proc.columns:
    df_proc = df_proc.drop(columns=["customerID"])

cat_cols = df_proc.select_dtypes(include=["object"]).columns
df_final = pd.get_dummies(df_proc, columns=cat_cols, drop_first=True)

df_final = df_final.fillna(df_final.mean(numeric_only=True))

X = df_final.drop(columns=["Churn_num"])
y = df_final["Churn_num"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train_scaled, y_train)
y_pred_log = log_reg.predict(X_test_scaled)

print("=== Regresión Logística ===")
print("Accuracy:", accuracy_score(y_test, y_pred_log))
print(confusion_matrix(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))

tree = DecisionTreeClassifier(random_state=42, max_depth=5)
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)

print("\n=== Árbol de Decisión ===")
print("Accuracy:", accuracy_score(y_test, y_pred_tree))
print(confusion_matrix(y_test, y_pred_tree))
print(classification_report(y_test, y_pred_tree))
```
